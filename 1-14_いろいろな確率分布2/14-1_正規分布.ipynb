```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
import math
```

```python
def normdist(x, mu, sigma):
    """
    正規分布関数
    x : 確率変数
    mu : 平均
    sigma : 標準偏差
    """
    coef = 1/((math.sqrt(2*math.pi)) * sigma) # 係数
    exponent = (-1) * (pow((x-mu), 2)) / (2 * pow(sigma, 2)) # 指数
    ret = coef * pow(math.e, exponent)
    return ret
```

```python
##### パラメータを定義 #####
mu = 10
sigma = 25
plt_min = -100
plt_max = 100
##########################

Figure, ax = plt.subplots() #グラフを定義

x = []
y = []
for i in range(plt_min, plt_max):
    x.append(i)
    y.append(normdist(i, mu, sigma))

ax.plot(x,y) #プロット
plt.show() #グラフを出力
Figure.savefig('14-1.jpg', dpi=100) #グラフを保存（dpiで解像度指定）
```
