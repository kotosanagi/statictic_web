```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
import math
```

```python
def normdist(x, mu, sigma):
    """
    正規分布関数
    x : 確率変数
    mu : 平均
    sigma : 標準偏差
    """
    coef = 1/((math.sqrt(2*math.pi)) * sigma) # 係数
    exponent = (-1) * (pow((x-mu), 2)) / (2 * pow(sigma, 2)) # 指数
    ret = coef * pow(math.e, exponent)
    return ret
```

```python
def std_normdist(x):
    """
    標準正規分布関数
    x : 確率変数
    mu : 平均 = 0
    sigma : 標準偏差 = 1
    """
    coef = 1/((math.sqrt(2*math.pi)) * sigma) # 係数
    exponent = (-1) * (pow(x, 2)) / 2 # 指数
    ret = coef * pow(math.e, exponent)
    return ret
```

```python
# 数学　平均点：60点　標準偏差：4点

##### パラメータを定義 #####
mu_1 = 60
sigma_1 = 4
plt_min = 0
plt_max = 100
##########################

Figure, ax = plt.subplots() #グラフを定義

x = []
y = []
for i in range(plt_min, plt_max):
    x.append(i)
    y.append(normdist(i, mu_1, sigma_1))

ax.plot(x,y) #プロット
plt.show() #グラフを出力
```

```python
# 国語　平均点：40点　標準偏差：5点

##### パラメータを定義 #####
mu_2 = 40
sigma_2 = 5
plt_min = 0
plt_max = 100
##########################

Figure, ax = plt.subplots() #グラフを定義

x = []
y = []
for i in range(plt_min, plt_max):
    x.append(i)
    y.append(normdist(i, mu_2, sigma_2))

ax.plot(x,y) #プロット
plt.show() #グラフを出力
```

***正規分布の再生性***  
このクラスの数学および国語の点数はそれぞれ異なる（独立な）正規分布に従うとき、数学と国語の点数を足した点数は、  
平均が  
mu_1 + mu_2 = 60 + 40 = 100  
分散が  
(sigma_1)^2 + (sigma_2)^2 = 4^2 + 5^2 = 41  
の正規分布に従うとのこと  

```python
# 数学＋国語　平均点：100点　標準偏差：41点

##### パラメータを定義 #####
mu = 100
sigma = math.sqrt(41) # 分散を標準偏差に変換
plt_min = 50
plt_max = 150
##########################

Figure, ax = plt.subplots() #グラフを定義

x = []
y = []
for i in range(plt_min, plt_max):
    x.append(i)
    y.append(normdist(i, mu, sigma))

ax.plot(x,y) #プロット
plt.show() #グラフを出力
```

***標準正規分布***

```python
# 標準正規分布

##### パラメータを定義 #####
plt_min = -3
plt_max = 3
separate = 10 # グラフを滑らかにする（値が大きいほど滑らかになる）
##########################

Figure, ax = plt.subplots() #グラフを定義

x = []
y = []
for i in range(plt_min*separate, plt_max*separate):
    x.append(i/separate)
    y.append(std_normdist(i/separate))

ax.plot(x,y) #プロット
plt.show() #グラフを出力
```
